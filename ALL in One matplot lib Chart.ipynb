```python
import csv 
x=[]
y=[]
with open('data science jay/nba.csv','r') as file:
    data=csv.reader(file)
    for col in data:
        x.append(col[4])# age
        y.append(col[6])# weight
plt.plot(x,y,label='analysis',color='orange')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("plot the csv data")
plt.legend()
plt.show()


        
```
